If divisible by 3, print Fizz.  
If divisible by 5, print Buzz.  
If divisible by both, print FizzBuzz.  
Else, print the number.

```python
# creating dictionary through comprehension
# not really crucial for this example

objects = ['blue', 'apple', 'dog']
categories = ['color', 'fruit', 'pet']
a_dict = {key: value for key, value in zip(categories, objects)}
a_dict
```

```python
import time
start_time = time.time()

for i in range(1,100+1):
    n = True           # print number
    d = {3:'Fizz',     # words to print in place of number
         5:'Buzz'}
    
    for num in d.items():
        if i%num[0] == 0:
            print(num[1],end='')
            n = False  # don't print number if printing word
    if n:
        print(i)       # if not printing word, print number and return line
    else:
        print()        # else just return line
        
print(f'\nElapsed Time: {time.time()-start_time:0.10f} seconds')
```

```python
start_time = time.time()
N = 100
d = {
    3:'Fizz',
    5:'Buzz'
}

for i in range(1,N+1):
    output = ''
    for divisor,word in d.items():
        if (i % divisor) == 0:
            output += word
    print(output or i)

print(f'\nElapsed Time: {time.time()-start_time:0.10f} seconds')
```
